```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
from sklearn.linear_model import LogisticRegression
# Generacin de datos para dos clases
np.random. seed (0)
n_samples = 200 # nmero de muestras por clase
class_1 = 0.6 * np. random. randn(n_samples, 2) # generacin de 200 muestras aleatorias bidimensionales (2D)
class_2 = 1.2 * np. random.randn(n_samples, 2) + np.array([5, 1]) # generacin de 200 muestras aleatorias bidimensionales
# (2D) pero con corrimiento
X = np. vstack((class_1, class_2)) #apila verticalmente (concatenar hacia abajo) dos conjuntos de datos
y = np.hstack((np.zeros(n_samples), np.ones(n_samples)))
# crea un vector de ceros con una longitud igual a
#n_samples para la clase 0 y clas 1 respectivamente
# Visualizacin de los datos
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.RdBu)
plt.xlabel('Caracterstica 1')
plt.ylabel('Caracterstica 2')
plt.title('Conjunto de Datos de Dos Clases')
plt.show ()
```

```python
# Paso 2: Encontrar los centros de las neuronas RBF mediante K-Means
n_clusters = 10
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
centers = kmeans.cluster_centers_
```

```python
# Paso 3: Calcular el ancho de los kernels RBF
widths = pairwise_distances_argmin_min(X, centers)[1].mean()
```

```python
# Paso 4: Calcular las activaciones de las neuronas RBF
def rbf_kernel(x, centers, width):
  return np.exp(-0.5 * (np.linalg.norm(x[:, np.newaxis] - centers, axis=2) / width) ** 2)
# x[:, np.newaxis] es una operacin que cambia la forma de la matriz de caractersticas
# Esto se hace para poder restar los centros de las neuronas centers, que tiene la forma (n_clusters, n_features)
# np. linalg-norm(x[:, np.newaxis] - centers, axis=2) calcula la norma Euclidiana
# entre cada punto de datos en x y todos los centros de las neuronas.
phi = rbf_kernel(X, centers, widths)
# Convertir phi a una matriz 2D
phi = phi. reshape(-1, n_clusters)
# convertimos la matriz phi a una matriz 2D utilizando .reshape(-1, n_clusters)
```

```python
# Paso 5: Entrenar un modelo de clasificacin en las activaciones de las neuronas RBF
rbf_model = LogisticRegression()
rbf_model. fit (phi, y)
```

```python
# Generar nuevos datos de prueba
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
# se calculan como los valores mnimos y mximos en la columna 0
# (Caracterstica 1) de los datos originales X, con un pequeo ajuste
# hacia abajo y hacia arriba respectivamente.
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np. meshgrid(np. arange(x_min, x_max, 0.1), np.arange (y_min, y_max, 0.1))
# se crean utilizando np.meshgrid para crear una malla de puntos en el espacio de caractersticas
Z = rbf_model.predict(rbf_kernel(np.c_[xx.ravel(), yy.ravel()], centers, widths))
# Se utiliza el modelo de regresin logstica rbf_model previamente entrenado para
# predecir las clases de los nuevos datos generados. Para hacerlo, se calcula el kernel
# RBF para cada punto en la malla (np.c_[xx.ravel(), yy.ravel()]) utilizando la funcin
#rbf_kernel. Luego, se pasa esta matriz de caractersticas a rbf._model. predict para
# obtener las predicciones de clase. Las predicciones se almacenan en la variable Z.
# Visualizar la frontera de decisin
Z = Z. reshape(xx.shape)
plt. contourf(xx, yy, Z, alpha=0.4)
plt. scatter (X[:, 0], X[:, 1], c=y, cmap=plt. cm.RdBu)
plt. xlabel ('Caracterstica 1')
plt. ylabel('Caracterstica 2')
plt. title('Clasificacin con una RBF')
plt. show()
```
