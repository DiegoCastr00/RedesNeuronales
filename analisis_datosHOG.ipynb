```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

# Asumiendo que X son tus características y y son tus etiquetas
X = df.drop('etiqueta', axis=1)
y = df['etiqueta']

modelo = RandomForestClassifier()

# Realizar la validación cruzada k-fold
scores = cross_val_score(modelo, X, y, cv=10)

print(f'Scores de validación cruzada: {scores}')
print(f'Promedio de scores: {scores.mean()}')
```
