```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import mahotas
import pandas as pd

ruta_carpeta = "dataset"
imagenes = []

for archivo in os.listdir(ruta_carpeta):
    ruta_archivo = os.path.join(ruta_carpeta, archivo)
    if os.path.isfile(ruta_archivo) and archivo.lower().endswith(('.jpg', '.jpeg', '.png')):
        imagenes.append(ruta_archivo)
        
print(len(imagenes))
```

```python
etiquetas = []
df = pd.DataFrame()
for ruta_imagen in imagenes:

    imagen = cv2.imread(ruta_imagen)
    imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

    radius = 15
    zernike = mahotas.features.zernike_moments(imagen, radius)
    
    df = pd.concat([df, pd.DataFrame(zernike).T], ignore_index=True)
    
    nombre_archivo = os.path.basename(ruta_imagen)

    etiqueta = nombre_archivo.split('_')[-1].split('.')[0]
    etiquetas.append(etiqueta)
columnas = ['zernike_' + str(i) for i in range(len(df.columns))]
df.columns = columnas
df['etiqueta'] = etiquetas
df
```

```python
df.to_csv('Zernike.csv', index=False)
```

```python
from skimage.feature import hog

etiquetas = []
HOG = pd.DataFrame()
for ruta_imagen in imagenes:

    imagen = cv2.imread(ruta_imagen)
    imagen = cv2.cvtColor(imagen, cv2.COLOR_RGB2GRAY)

    hog_descriptor = hog(imagen, orientations=8, pixels_per_cell=(16, 16),
                        cells_per_block=(1, 1), visualize=False)

    HOG = pd.concat([HOG, pd.DataFrame(hog_descriptor).T], ignore_index=True)
    
    nombre_archivo = os.path.basename(ruta_imagen)
    etiqueta = nombre_archivo.split('_')[-1].split('.')[0]
    etiquetas.append(etiqueta)
    
columnas = ['HOG_' + str(i) for i in range(len(HOG.columns))]
HOG.columns = columnas
HOG['etiqueta'] = etiquetas
HOG
```

```python
HOG.to_csv('HOG.csv', index=False)
```

```python
from skimage.feature import hog

etiquetas = []
HOG_cv2 = pd.DataFrame()
for ruta_imagen in imagenes:

    imagen = cv2.imread(ruta_imagen)
    imagen = cv2.cvtColor(imagen, cv2.COLOR_RGB2GRAY)
    
    cell_size = (16, 16)
    win_size = (imagen.shape[1] // cell_size[1] * cell_size[1], imagen.shape[0] // cell_size[0] * cell_size[0])
    block_size = (2 * cell_size[1], 2 * cell_size[0])
    block_stride = cell_size
    num_bins = 4
    
    hog = cv2.HOGDescriptor(win_size, block_size, block_stride, cell_size, num_bins)
    
    hog_descriptor = hog.compute(imagen)
    
    HOG_cv2 = pd.concat([HOG_cv2, pd.DataFrame(hog_descriptor).T], ignore_index=True)
    
    nombre_archivo = os.path.basename(ruta_imagen)
    etiqueta = nombre_archivo.split('_')[-1].split('.')[0]
    etiquetas.append(etiqueta)
    
columnas = ['HOG_' + str(i) for i in range(len(HOG_cv2.columns))]
HOG_cv2.columns = columnas
HOG_cv2['etiqueta'] = etiquetas
HOG_cv2
```

```python
HOG_cv2.to_csv('HOG_TODO.csv', index=False)
```
